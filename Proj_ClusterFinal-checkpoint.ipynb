```python
#Análise DF1
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler ()
scaled_df1 = scaler.fit_transform (df1)
scaled_df1
```

```python
scaled_df1.shape
```

```python
#Criando uma função para achar o melhor número de 'cluster' para o exemplo

from sklearn.cluster import KMeans

def get_kmeans(numero_clusters, scaled_df1):
    
    model = KMeans (n_clusters=numero_clusters)
    model.fit(scaled_df1)
    return [numero_clusters, model.inertia_]  
```

```python
resultado = [get_kmeans(numero_de_grupos , scaled_df1) for numero_de_grupos in range (1,15)]
resultado
```

```python
resultado= pd.DataFrame(resultado, columns=['groups', 'inertia'])
```

```python
resultado.head()
```

```python
resultado['inertia'].plot(xticks=resultado['groups'], subplots=True, figsize = (30,10), sharex=False, rot=1)
```

```python
# Df1 = 4 clusters
```

```python
#Pelo método do cotovelo, vemos que a primeira grande variação na angulação à esquerda é dada em 14. 
#Logo esse é o número de clusters indicado

from sklearn.cluster import KMeans

model = KMeans (n_clusters = 4)
model.fit(scaled_df1)
y=model.labels_
```

```python
df1[model.labels_== 2]
```

```python
df_groups1= pd.DataFrame(model.cluster_centers_,columns=df1.columns)
df_groups1.head()
```

```python
df_groups1.transpose()
```

```python
df_groups1.transpose().plot.bar(subplots=True, figsize = (30,20), sharex=False, rot=1)
```

```python
pd.DataFrame(y)[0].value_counts()
```
